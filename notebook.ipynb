```python
import imageio
import matplotlib.pyplot as plt

from filters.bilateral import filter as bilateral
from filters.unsharp import filter as unsharp
```

```python
plt.rcParams["figure.figsize"] = (20,10)

camera = imageio.imread("images/camera.png")
moon = imageio.imread("images/moon.png")
arara = imageio.imread("images/arara.png")
```

```python
def show(original, modified):
    fig, ax = plt.subplots(1, 2)
    ax[0].imshow(original)
    ax[1].imshow(modified)
```

```python
show(camera, bilateral(camera, 3, 150.0, 100.0))
```

```python
show(moon, bilateral(moon, 5, 50.0, 100.0))
```

```python
show(arara, unsharp(arara, 0.5, 1))
```

```python
show(arara, unsharp(arara, 0.75, 2))
```
